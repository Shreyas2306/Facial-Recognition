```python
#Importing Libraries
import cv2
import sys


#imagePath = sys.argv[1]

#Importing haarcascade_frontalface_default.xml
cascPath = './haarcascade_frontalface_default.xml'
faceCascade = cv2.CascadeClassifier(cascPath)

#Initialing Video Capturing for capturing faces and recognising them
video_capture = cv2.VideoCapture(0)
rec =cv2.face.LBPHFaceRecognizer_create()

#Reading the captured images for recognition  
rec.read('./trainingdata.yml')

#Initialising id to 0
id=0
fontface = cv2.FONT_HERSHEY_SIMPLEX
fontscale = 1
fontcolor = (255, 255, 255)

while True:
    
    # Capture frame-by-frame
    ret, frame = video_capture.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray, 1.3, 5)
    for (x, y, w, h) in faces:
        
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        Id, conf = rec.predict(gray[y:y+h,x:x+w])
        cv2.putText(frame,str(Id), (x,y+h),fontface, fontscale, fontcolor)
        
        cv2.waitKey(1)
        cv2.destroyAllWindows()

    # Display the resulting frame
        cv2.imshow('faces', frame)
        cv2.waitKey(1)
    
        video_capture.release()
        cv2.destroyAllWindows()
    if (cv2.waitKey(1)== ord('q') ):
        break
            
        
    

# When everything is done, release the capture
video_capture.release()
cv2.destroyAllWindows()
cv2.waitKey(100)
cv2.destroyAllWindows()


```
